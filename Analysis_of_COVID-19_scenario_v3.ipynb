It is possible to see thta there are continent features in this dataset. Let's drop them.

```
top5_total = df.loc[df.date == "2021-03-24", ["location", "date", "total_deaths", "total_cases"]]

filt = (top5_total[(top5_total["location"] == "South America") | (top5_total["location"] == "Africa") | (top5_total["location"] == "Asia")
        | (top5_total["location"] == "North America") | (top5_total["location"] == "World") | (top5_total["location"] == "European Union")
         | (top5_total["location"] == "Europe")])

top5_total.drop(filt.index, inplace=True)

top5_total.sort_values(by="total_deaths", ascending=False).head(10)
```

That's a good first approach to know more about the Covid-19 scenario in the world. However, there is a better way to display this information. Let's plot it! To do so, let's use `Matplotlib` library. Let's first plot a histogram of the total number of cases and deaths.

Seaborn library has only been imported to better Matplotlib layout (`sns.set()`).

```
# importing Matplotlib

import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

# creating a graph of total number of cases
sns.set()

top5_total_deaths_chart = top5_total.loc[top5_total.date == "2021-03-24", ["location", "total_deaths"]].sort_values(by="total_deaths", ascending=False)[0:5]
top5_total_cases_chart = top5_total.loc[top5_total.date == "2021-03-24", ["location", "total_cases"]].sort_values(by="total_cases", ascending=False)[0:5]


fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,7))
ax1.bar(x=top5_total_deaths_chart["location"], height=top5_total_deaths_chart["total_deaths"], facecolor="r", alpha=0.75)
ax1.set_title("Total deaths", fontsize = 20)
ax1.set_ylabel("Deaths", fontsize=15, labelpad=10, )
ax1.grid(ls="--")
ax1.tick_params(labelsize=12)


ax2.bar(x="location", height="total_cases", data = top5_total_cases_chart)
ax2.set_title("Total cases", fontsize=20)
ax2.set_ylabel("Cases", fontsize=15, labelpad=10)
ax2.grid(ls="--")
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
ax2.tick_params(labelsize=12)


fig.autofmt_xdate()

plt.tight_layout()
```

### Analyzing the evolution of Covid-19 cases and deaths in the world

```
#Obtaining part of the dataset to be used in the world analysis

cases_world = df.loc[df.location == "World", ["date","total_cases"]]
deaths_world = df.loc[df.location=="World", ["date","total_deaths"]]
```

```
#Line plot for total deaths and cases in the world

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(ncols=2, nrows=2, figsize=(18,5))


ax1.plot(cases_world["date"], cases_world["total_cases"], linewidth=3, color=None, linestyle="solid")
ax1.text(x=cases_world.iloc[-1,0] + dt.timedelta(days=2), y=cases_world.iloc[-1,1], s=f'{int(cases_world.iloc[-1,1])}', alpha=0.7)
ax1.set_title("Covid-19 - World Total Cases - Regular Scale")
ax1.set_xlim(cases_world.iloc[0,0], cases_world.iloc[-1,0] + dt.timedelta(days=38))

ax2.plot(cases_world["date"], cases_world["total_cases"], linewidth=3, color=None, linestyle="solid")
ax2.set_yscale("log")
ax2.text(x=cases_world.iloc[-1,0] + dt.timedelta(days=2), y=cases_world.iloc[-1,1], s=f'{int(cases_world.iloc[-1,1])}', alpha=0.7)
ax2.set_title("Covid-19 - World Total Cases - Logarithmic Scale")
ax2.set_xlim(cases_world.iloc[0,0], cases_world.iloc[-1,0] + dt.timedelta(days=38))

ax3.plot(deaths_world["date"], deaths_world["total_deaths"], linewidth=3, color="red", linestyle="solid")
ax3.text(x=deaths_world.iloc[-1,0] + dt.timedelta(days=2), y=deaths_world.iloc[-1,1], s="{}".format(int(deaths_world.iloc[-1,1])) , alpha=0.7)
ax3.set_xlim(deaths_world.iloc[0,0], deaths_world.iloc[-1,0] + dt.timedelta(days=38))
ax3.set_title("Covid-19 - World Total Deaths - Regular Scale")
plt.tight_layout()

ax4.plot(deaths_world["date"], deaths_world["total_deaths"], linewidth=3, color="red", linestyle="solid")
ax4.set_yscale("log")
ax4.text(x=deaths_world.iloc[-1,0] + dt.timedelta(days=2), y=deaths_world.iloc[-1,1], s="{}".format(int(deaths_world.iloc[-1,1])) , alpha=0.7)
ax4.set_xlim(deaths_world.iloc[0,0], deaths_world.iloc[-1,0] + dt.timedelta(days=38))
ax4.set_title("Covid-19 - World Total Deaths - Logarithmic Scale")
plt.tight_layout()

plt.show()
```

Because our data varies a lot, the number of cases and deaths of Covid-19 has been represented in a logarithmic scale. This result shows that the increase of death follows suit the increase of cases. February 2020 presented higher increase of cases and deaths. It may be related to the beginning of the Coronavirus outbreak when nobody had idea of how serious the desease was and how to treat it. Let's now see how the the evolution of the number of cases and deaths along the year for these countries. To accomplish this task let's consider a 7-day rolling average graph. 

The first thing is to create a dataframe containing total data from the world and the other countries. Let's do it.

```
# Choosing the columns that matters to us

col = ["location","date", 'total_cases', 'total_deaths', 'new_cases', 'new_deaths']

# Creating the DataFrames 

df_world = df.loc[df["location"]=="World", col]
df_usa = df.loc[df["location"]=="United States", col]
df_brazil = df.loc[df["location"] == "Brazil", col]
df_india = df.loc[df["location"] == "India", col]
df_mexico = df.loc[df["location"]=="Mexico", col]

#Merging the DataFrames

df_countries = df_world.merge(df_usa, on="date", suffixes=("_world","_usa"))

df_countries_2 = df_countries.merge(df_brazil, on="date", suffixes=(None, "_brazil"))

df_countries_3 = df_countries_2.merge(df_india, on="date", suffixes=(None, "_india"))

df_countries_4 = df_countries_3.merge(df_mexico, on="date", suffixes=(None, "_mexico"))

df_countries_4 = df_countries_4.rename(columns={"total_cases":"total_cases_brazil", "total_deaths":"total_deaths_brazil", "new_cases":"new_cases_brazil", "new_deaths":"new_deaths_brazil"})


#dropping useless columns

df_countries_4.drop(labels=["location_world", "location_usa", "location", "location_india", "location_mexico"], axis=1, inplace=True)

df_countries_4
```

Great ... The following steps are summarized below:


* To calculate each country's participation for each column;

* To fill the null values left by divisions by zero with zero;

* To drop the columns that are not usable anymore;

* To apply an equation to transform each column a 7-day average;

* To fill null values after  calculating the moving average with zero.
