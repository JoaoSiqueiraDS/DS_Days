```
# Knowing how many people know Python in each country

filt1 = df["Country"] == "Brazil"
df.loc[filt1]['LanguageWorkedWith'].str.contains("Python").sum()
```

```
uses_python = country_grp["LanguageWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
uses_python
```

```
df_respondents = df["Country"].value_counts()

df_respondents
```

```
python_df = pd.concat([df_respondents, uses_python], axis="columns", sort=False)

python_df
```

```
python_df.rename(columns={"Country":"NumRespondents", "LanguageWorkedWith":"Python_Users"}, inplace=True)

python_df
```

```
python_df["pctKnowPython"] = (python_df["Python_Users"]/python_df["NumRespondents"])*100

python_df.loc["Brazil"]
```
