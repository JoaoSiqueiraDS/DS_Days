# 5.1.2 Generating example of regression data

```
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt


n = 100
beta_0 = 5
beta_1 = 2
np.random.seed(2)
x = 10 * ss.uniform.rvs(size=n)
y = beta_0 + beta_1 * x + ss.norm.rvs(loc=0, scale=1, size=n)

plt.figure()
plt.plot(x, y, "o", ms=5)
xx = np.array([0,10])
plt.plot(xx, beta_0 + beta_1 * xx)
plt.xlabel("x")
plt.ylabel("y")
```

# 5.1.3 - Simple Linear Regression

The goal is to predict a quantitative response Y based on a predictor X. It is assumed the following relationship between the random variables X and Y.

Y = Beta0 + Beta1.X + Epsilon (All capital letters Y and X correspond to random values). 

Once we have used trainning data to produce estimates `Beta0hat` and `Beta1hat` for the modal coefficients **we can predict future values of Y.**

yhat = Beta0hat + Beta1hatx

yhat indicates a prediction of the random value Y

The hats in the Betas mean that they have been estimated through the use of data.

* Most common approach to estimate the modal parameters involves minimizim the `least squares criterion`. 

* Residue and RSS: Modal coefficients (Betas) are the ones that minimizes the RSS criterion.
