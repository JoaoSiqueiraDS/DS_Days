# 5.1.1 - Introduction to statistical learning

It is divided into supervised learning and unsurpervised learning.

* Supervised learning - inputs and outputs ... Outputs provided by a supervisor

* Unsurpervised learning - inputs only - The goal here is to learn relationship structure from such data


Supervised learning: 

* Regression: continous outputs
* Classification: Categorical outputs

Variables in statistics can be: 

* Quantitative: such as income
* Qualitative: such as female and male

* Quantitative: regression problems
* Qualitative: classification problems

It's common to involve both quantitative and qualitative varialbles in regression and classification problems.
